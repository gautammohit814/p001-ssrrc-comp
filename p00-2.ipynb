# Data manipulation with Numpy and Pandas



```python
import numpy as np
import pandas as pd
```

```python
# read csv file
cars = pd.read_csv('read_data.csv')
cars.head()
```

```python
# list all columns
cars.columns
```

```python
cars.keys()
```

```python
cars.index = cars.pop('car')
cars.head()
```

```python
# describe dataset
cars.describe()
```

```python
cars.sort_index(inplace=True)
cars.head(3)
```

```python
cars.sort_index(inplace=True)
cars.head(100).tail(5)
```

```python
cars.sort_values('mpg').head(15)
```

```python
cars.sort_values('hp', ascending=False).head()
```

## Selection

**Note**:  pandas-specific data accessor methods, `.at`, `.iat`, `.loc`, `.iloc` and `.ix`.


```python
cars['mpg']
```

```python
# slice of rows
cars[2:5]
```

```python
cars[5:]
```

```python
cars.loc['Fiat 128':'Lotus Europa']
```

```python
# selection by position
cars.iloc[3]
```

```python
cars.iloc[3:5, 0:2]
```

```python
cars[cars.cyl > 6] 
```

```python
#stats
cars.mean()
```

```python
cars.mean(axis=1)
```

```python
cars.groupby('cyl').mean()
```
